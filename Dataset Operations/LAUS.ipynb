```python
import pandas as pd

dfs = []
for yr in range(2010, 2018):
    path = f"Raw_Datasets/LAUS/laucnty{str(yr)[2:]}.xlsx"
    df_yr = pd.read_excel(path, skiprows=1, header=0)
    df_yr['Year'] = yr-2000
    df_yr["State FIPS"] = df_yr["State FIPS Code"]
    df_yr["County FIPS"] = df_yr["County FIPS Code"]
    dfs.append(df_yr)

laus = pd.concat(dfs, ignore_index=True)
```
