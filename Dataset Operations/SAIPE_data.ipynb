```python
import pandas as pd


PATH_TO_SAIPE = "Raw_Datasets/SAIPE"
get_yearly_sapie_datafile_name = lambda x: f'{PATH_TO_SAIPE}/est{x}all.xls'
```

```python
get_yearly_sapie_datafile_name(11)
```

```python
YEARS_AVAILABLE = [x for x in range(10,18)]
dfs = []

for yr in YEARS_AVAILABLE:
    yearly_df = pd.read_excel(get_yearly_sapie_datafile_name(11), header=2)
    yearly_df["Year"] = yr
    yearly_df = yearly_df.iloc[:-3,:]
    dfs.append(yearly_df)

saipe = pd.concat(dfs)
saipe
```

```python
saipe.to_csv("Cleaned_Datasets/SAIPE.csv", index=False)
```
