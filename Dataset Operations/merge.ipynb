```python
import pandas as pd


saipe = pd.read_csv("Cleaned_Datasets/SAIPE.csv")
snap = pd.read_csv("Cleaned_Datasets/SNAP.csv")
mmg = pd.read_csv("Cleaned_Datasets/MMG.csv")
```

```python
saipe["State"] = saipe["Postal"]
saipe
```

```python
snap
```

```python
mmg
```

```python
merged = pd.merge(saipe,snap, on=["State", "Name", "Year"], how="inner")
merged
```

```python
merged.Year.value_counts()
```

```python
merged = pd.merge(merged,mmg, on=["State", "Name", "Year"], how="inner")
merged
```

```python
merged.Year.value_counts()
```

```python
laus = pd.read_csv("Cleaned_Datasets/LAUS.csv", index_col=0)
laus
```

```python
merged = pd.merge(merged, laus, on=["State FIPS", "County FIPS", "Year"], how="inner")
merged
```
